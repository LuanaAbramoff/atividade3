### Comparação - Precisão

```python
import matplotlib.pyplot as plt
import numpy as np

# Dados das entradas
entradas = ['Entrada 1', 'Entrada 2', 'Entrada 3', 'Entrada 4']

# Valores máximos alcançados por cada tarefa
valores_max_t1 = [334, 283, 3320, 463]  # Tarefa 1
valores_max_t2 = [285, 282, 1948, 464]  # Tarefa 2

# Valores ideais alcançados pela busca exaustiva
valores_ideais = [601, 870, 3520, 648]

# Configurações do gráfico
bar_width = 0.35
index = np.arange(len(entradas))

# Criação das barras
fig, ax = plt.subplots(figsize=(10, 6))

bar1 = ax.bar(index, valores_max_t1, bar_width, label='Embaralhamento', color='b')
bar2 = ax.bar(index + bar_width, valores_max_t2, bar_width, label='Sel. Aleatória', color='r')

# Adicionando a linha representando o valor ideal
for i in range(len(valores_ideais)):
    ax.plot([i - bar_width / 2, i + bar_width * 1.5], [valores_ideais[i], valores_ideais[i]], 'g--', linewidth=2, label='Valor Ideal' if i == 0 else "")

# Configurações dos eixos
ax.set_xlabel('Entradas')
ax.set_ylabel('Valor Alcançado (dinheiros)')
ax.set_title('Comparação do Valor Máximo Alcançado por Tarefa')
ax.set_xticks(index + bar_width / 2)
ax.set_xticklabels(entradas)

# Adicionando a legenda
ax.legend()

# Mostrando o gráfico
plt.tight_layout()
plt.show()

```

Comparando, em termos de precisão, as técnicas de embaralhamento e de seleção aletória baseada em probabilidade, é possível concluir que as duas tiveram desempenhos bem próximos, com a técnica de embaralhamento sendo ligeiramente melhor. O que chama atenção é que ambas as técnicas de heurísticas aleatórias apresentam suas melhores soluções, entre as 5 execuções para cada uma das entradas, sendo muito piores do que a solução chamada de ideal, encontrada pela busca exaustiva. Com isso, é possível ver que a precisão quanto ao melhor resultado é uma grande desvantagem dos algoritmos com heurísticas aleatórias em relação ao algoritmo de busca exuastiva.

### Comparação entre as duas heurísticas em relação ao tempo e maior valor obtido

```python
import matplotlib.pyplot as plt

# Dados das entradas
entradas = ['Entrada 1', 'Entrada 2', 'Entrada 3', 'Entrada 4']

# Tempos de execução e valores alcançados
tempos_execucao_t1 = [[0.0004467, 0.0004595, 0.0003936, 0.0005753, 0.0004624],
                      [0.0006793, 0.0004632, 0.0005805, 0.0005846, 0.0008896],
                      [0.000471, 0.000471, 0.000471, 0.000471, 0.000471],
                      [0.000396, 0.000451, 0.0005819, 0.0004035, 0.0007635]]

valores_alcancados_t1 = [[334, 273, 236, 264, 295],
                         [283, 263, 239, 255, 264],
                         [3320, 3320, 3320, 3320, 3320],
                         [463, 575, 409, 492, 524]]

tempos_execucao_t2 = [[0.0003946, 0.0006408, 0.0006322, 0.0004905, 0.0009196],
                      [0.0003265, 0.0006186, 0.0009367, 0.0004886, 0.000369],
                      [0.00053, 0.0006343, 0.0010004, 0.0004718, 0.0009374],
                      [0.0004287, 0.0009614, 0.0008447, 0.0008101, 0.0003538]]

valores_alcancados_t2 = [[285, 245, 252, 295, 276],
                         [212, 254, 152, 282, 258],
                         [1102, 1948, 1810, 1393, 1462],
                         [404, 464, 318, 259, 324]]


# Criação dos gráficos
fig, axs = plt.subplots(2, 2, figsize=(12, 10))

for i, ax in enumerate(axs.flat):
    # Gráfico de dispersão para Tarefa 1
    ax.scatter(tempos_execucao_t1[i], valores_alcancados_t1[i], color='b', label='Embaralhamento', s=100, marker='o')
    
    # Gráfico de dispersão para Tarefa 2
    ax.scatter(tempos_execucao_t2[i], valores_alcancados_t2[i], color='r', label='Seleção Aleatória', s=100, marker='x')
    
    
    # Configurações do gráfico
    ax.set_xlabel('Tempo de Execução (segundos)')
    ax.set_ylabel('Valor Alcançado (dinheiros)')
    ax.set_title(f'Comparação de Entradas - {entradas[i]}')
    ax.legend()
    ax.grid(True)

# Ajuste do layout
plt.tight_layout()
plt.show()
```

Esses gráficos mostram os tempos e os valores obtidos nas 5 execuções para cada uma das entradas nas duas tarefas. É possível ver que exceto no caso da Entrada 2, a heurística de embaralhamento consegue o maior valor em menor tempo, mostrando que é melhor nos dois sentidos. Na entrada 2, vemos que a seleção aleatória consegue um valor bem próximo do maior dos dois algoritmos, em menor tempo. É interessante observar que no caso da entrada 3, as 5 execuções com a heurística de embaralhamento obtiveram o mesmo valor no mesmo tempo. Isso aconteceu por uma característica específica dessa entrada, que tem itens que ou tem o mesmo valor ou mesmo peso que outros, então quando ocorre o embaralhamento, o valor resultante tem uma grande chance de ser o mesmo do anterior.

### Gráfico - Tamanho das Entradas x Precisão

```python
import matplotlib.pyplot as plt

# Tamanhos das entradas
tamanhos_entradas = [30,50,75]

# Valores máximos alcançados por cada tarefa
valores_max_t1 = [575, 334, 283]  # Tarefa 1
valores_max_t2 = [464, 285, 282]  # Tarefa 2 

# Valores ideais alcançados pela busca exaustiva
valores_ideais = [648, 601, 870]

# Cálculo da precisão (razão entre valor máximo alcançado e valor ideal)
precisao_t1 = [val_max / val_ideal for val_max, val_ideal in zip(valores_max_t1, valores_ideais)]
precisao_t2 = [val_max / val_ideal for val_max, val_ideal in zip(valores_max_t2, valores_ideais)]

# Criação do gráfico
plt.figure(figsize=(10, 6))

plt.plot(tamanhos_entradas, precisao_t1, 'bo-', label='Embaralhamento', markersize=8)
plt.plot(tamanhos_entradas, precisao_t2, 'ro-', label='Seleção aleatória', markersize=8)

# Configurações do gráfico
plt.xlabel('Tamanho das Entradas')
plt.ylabel('Precisão')
plt.title('Precisão das Tarefas em Relação ao Tamanho das Entradas')
plt.legend()
plt.grid(True)

# Mostrando o gráfico
plt.tight_layout()
plt.show()
```

Esse gráfico foi feito para ver qual é o efeito do tamanho das entradas na precisão dos algoritmos com heurísticas aleatórias. A precisão foi calculada com uma razão entre o valor máximo obtido nas eurísticas para cada entrada e o valor obtido na busca exaustiva. É possível ver que a precisão nas duas heurísticas decai fortemente quando o tamanho das entradas aumentam. Além disso a diferença de precisão entre a heurística de embaralhamento e a heurística de seleção aleatória cai de acordo com o tamanho das entradas.
